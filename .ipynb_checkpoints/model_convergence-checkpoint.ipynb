```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import os

exp_dir = './saved_models_MLP'
exp_ib1 = os.path.join(exp_dir, 'exp-ib1-fcc.csv')
exp_ib3 = os.path.join(exp_dir, 'exp-ib3-fcc.csv')
exp_f = os.path.join(exp_dir, 'exp-f-fcc.csv')
exp_onehot = os.path.join(exp_dir, 'exp-onehot-fcc.csv')

exp_ib1_bcc = os.path.join(exp_dir, 'exp-ib1-bcc.csv')
exp_ib3_bcc = os.path.join(exp_dir, 'exp-ib3-bcc.csv')
exp_f_bcc = os.path.join(exp_dir, 'exp-f-bcc.csv')
exp_onehot_bcc = os.path.join(exp_dir, 'exp-onehot-bcc.csv')

exp_f_eform_fcc = os.path.join(exp_dir, 'exp-f-eform-fcc.csv')
exp_ib1_eform_fcc = os.path.join(exp_dir, 'exp-ib1-eform-fcc.csv')
exp_ib3_eform_fcc = os.path.join(exp_dir, 'exp-ib3-eform-fcc.csv')

```

```python
def merge_exp_csv(file_list, features=['ib1', 'ib3']):
    # The following settings are based on your dataset.
    rows = 5
    fractions = [0.2, 0.4, 0.6, 0.8, 1]
    maes = []
    feature = []
    data_size = []
    for i, f in enumerate(file_list):
        df_i = pd.read_csv(f)
        feature_i = i
        for idx in range(rows):
            a = df_i.iloc[idx]
            ncols = len(a)
            maes.extend(a)
            feature.extend([features[feature_i] for _ in range(ncols)])
            data_size.extend([fractions[idx] for _ in range(ncols)])

    merged_data = pd.DataFrame({'data_size': data_size,
                                'mae': maes,
                                'feature': feature})

    return merged_data
```

```python
# df = merge_exp_csv([exp_ib1, exp_ib3], features=['ib1', 'ib3'])
# df = merge_exp_csv([exp_ib1_bcc, exp_ib3_bcc], features=['ib1', 'ib3'])
df = merge_exp_csv([exp_ib1_bcc, exp_ib3_bcc, exp_f_bcc, exp_onehot_bcc], features=['IB$_1$', 'IB$_3$', 'f', 'onehot'])

df_fcc = merge_exp_csv([exp_ib1, exp_ib3, exp_f, exp_onehot], features=['IB$_1$', 'IB$_3$', 'f', 'onehot'])
# for idx, value in df.iterrows():
#     print(idx, value['exp_1'])
# print(df)

df_fcc_eform = merge_exp_csv([exp_ib3_eform_fcc, exp_f_eform_fcc], features=[ 'IB$_3$', 'f'])
```

```python
# magnetci property
sns.set_theme(style='darkgrid')
# fmri = sns.load_dataset(name='fmri')

sns.lineplot(x='data_size', y='mae', hue ='feature', data=df)
plt.show()
```

```python
# formation property
sns.set_theme(style='darkgrid')
# fmri = sns.load_dataset(name='fmri')

sns.lineplot(x='data_size', y='mae', hue ='feature', data=df_fcc_eform)
plt.show()
```

```python
plt.figure(figsize=(8, 4))

plt.subplot(121)
sns.set_theme(context='notebook', style='ticks', palette=None ) # style='white' ,one of {darkgrid, whitegrid, dark, white, ticks}

sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5}) # 设置context、字体大小以及其他参数

sns.set_style({"xtick.direction": "out"}) # 用dict将一系列控制参数赋值给axes_style()函数和set_style()函数的rc参数

# markers = {'ib1':'s', 'ib3':'o'}
pa = sns.lineplot(x='data_size', y='mae', markers = ['o', 's', '^', '<'], hue='feature', style ='feature', dashes = False, data=df_fcc)
pa.set_xlabel('Data size (%)')
pa.set_ylabel('MAE ($\mu_b$ / atom)')
h,l = pa.get_legend_handles_labels()
pa.legend(h, ['IB1', 'IB3', 'f', 'onehot'])
# pa.legend(['IB1', '', 'IB3', 'f', 'onehot'])
pa.text(-0.15, 1.05, '(a)', transform=pa.transAxes)

plt.subplot(122)
pb =sns.lineplot(x='data_size', y='mae', markers = ['o', 's', '^', '<'], hue='feature', style ='feature', dashes = False, data=df)
pb.set_xlabel('Data size (%)')
pb.set_ylabel('')
pb.legend(labels='', frameon=False)
pb.text(-0.15, 1.05, '(b)', transform=pb.transAxes)
# plt.show()
plt.savefig('./fig/convergence/fig1.png', dpi=500, bbox_inches='tight')
```
